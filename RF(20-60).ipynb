```python
from RF_for_BC import RF_for_BC
import numpy as np
import json
RF_model = RF_for_BC()
RF_model.load_model('./RF_models/models_20~60')
with open('./markers.json', 'r', encoding='utf-8') as f:
    markers = json.load(f)
target_gene = np.random.choice(markers)
# target_gene = '1_135401720'
target_gene = '3_230143954'
print(target_gene)
```

```python
# 未迭代
RF_model.load_model('./RF_models/models_20~60/')    
reward = 0.0
for i in range(30):
    seq = RF_model.generation_decision(target_gene)
    reward += seq[0]['Reward']
print(target_gene, 'Reward: ', reward / 30)
```

```python
# 迭代训练模型
RF_model.load_model('./RF_models/models_20~60/')
RF_model.reset_memory()
for i in range(1):
    RF_model.iteration_one_time(target_gene)
RF_model.save_model('./RF_models/models_20~60_after_iter/')
```

```python
# 迭代1轮的模型
RF_model.load_model('./RF_models/models_20~60_after_iter/')    
reward = 0.0
for i in range(30):
    seq = RF_model.generation_decision(target_gene)
    reward += seq[0]['Reward']
print(target_gene, 'Reward: ', reward / 30)
```
